```python
sns.set(style="whitegrid")

# Criar um grfico de linha da frequncia cardaca ao longo do tempo
plt.figure(figsize=(20, 5))  # Definir o tamanho da figura
sns.lineplot(x='timestamp', y='heart_rate', data=df, marker='o', color='red')

# Adicionando ttulo e labels
plt.title('Frequncia Cardaca ao Longo do Tempo')
plt.xlabel('Timestamp')
plt.ylabel('Frequncia Cardaca (bpm)')

# Exibir o grfico
plt.show()
```

```python
sns.set(style="whitegrid")

# Criar um grfico de linha da frequncia cardaca ao longo do tempo
plt.figure(figsize=(20, 5))  # Definir o tamanho da figura
sns.lineplot(x='timestamp', y='cadence', data=df, marker='o', color='gray')

# Adicionando ttulo e labels
plt.title('Cadncia ao Longo do Tempo')
plt.xlabel('Timestamp')
plt.ylabel('Cadncia (spm)')

# Exibir o grfico
plt.show()
```

```python
sns.set(style="whitegrid")

# Criar uma figura para o grfico
plt.figure(figsize=(10, 6))

# Plotar a rota de GPS
plt.plot(df['longitude'], df['latitude'], label='Rota GPS', color='blue', marker='.')

# Adicionando ttulo e labels com um estilo mais integrado
plt.title('Visualizao da Rota de GPS')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Ajustando a proporo dos eixos para evitar distoro
plt.axis('equal')

# Exibir o grfico
plt.show()
```
