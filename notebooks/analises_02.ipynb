```python
sns.set(style="whitegrid")

# Criar um gráfico de linha da frequência cardíaca ao longo do tempo
plt.figure(figsize=(20, 5))  # Definir o tamanho da figura
sns.lineplot(x='timestamp', y='heart_rate', data=df, marker='o', color='red')

# Adicionando título e labels
plt.title('Frequência Cardíaca ao Longo do Tempo')
plt.xlabel('Timestamp')
plt.ylabel('Frequência Cardíaca (bpm)')

# Exibir o gráfico
plt.show()
```

```python
sns.set(style="whitegrid")

# Criar um gráfico de linha da frequência cardíaca ao longo do tempo
plt.figure(figsize=(20, 5))  # Definir o tamanho da figura
sns.lineplot(x='timestamp', y='cadence', data=df, marker='o', color='gray')

# Adicionando título e labels
plt.title('Cadência ao Longo do Tempo')
plt.xlabel('Timestamp')
plt.ylabel('Cadência (spm)')

# Exibir o gráfico
plt.show()
```

```python
sns.set(style="whitegrid")

# Criar uma figura para o gráfico
plt.figure(figsize=(10, 6))

# Plotar a rota de GPS
plt.plot(df['longitude'], df['latitude'], label='Rota GPS', color='blue', marker='.')

# Adicionando título e labels com um estilo mais integrado
plt.title('Visualização da Rota de GPS')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Ajustando a proporção dos eixos para evitar distorção
plt.axis('equal')

# Exibir o gráfico
plt.show()
```
